```python
def cleanDate(dateToClean):
    with open('./../data/'+dateToClean) as json_data:
        d = json.loads(json_data.read())
        textList = [None] * len(d);
        count = 0;
        # TAGS THAT ARE REMOVED FROM TWEETS
        tags = ['http','https','pic.twitter','mined']
        for i in range (0, len(d)):
            finded = 0;
            for e in range (0,len(tags)):
                if d[i]['text'].find(tags[e]) != -1:
                    finded = 1;
            if finded != 1:
                textList[count] = d[i]['text'];
                count = count + 1;
        textList = textList[0:count];
        textList = json.dumps(textList);
        with open('./../cleanData/'+dateToClean, 'w') as f:
             json.dump(textList, f)
```

```python
import json;
import numpy as np;
import datetime;
import os;
import shutil;

years = ['2017','2018'];
dates = {'01':31,
       '02':28,
       '03':31,
       '04':30,
       '05':31,
       '06':30,
       '07':31,
       '08':31,
       '09':30,
       '10':31,
       '11':30,
       '12':31};
now = datetime.datetime.now();
currentDate = str(now.year)+'-'+str(now.month)+'-'+str(now.day)+'.json';

mypath = './../cleanData';
if not os.path.isdir(mypath):
    os.makedirs(mypath);
else:
    shutil.rmtree(mypath);
    os.makedirs(mypath);

for k in range(0,len(years)):
    for i in dates:
        for e in range(1,dates[i] + 1):
            if e < 10:
                temp = years[k]+'-'+str(i)+'-0'+str(e)+'.json'
            else:
                temp = years[k]+'-'+str(i)+'-'+str(e)+'.json'
            if temp == currentDate:
                break;
            try:
                cleanDate(temp);
            except json.JSONDecodeError as detail:
                print('Your File Has No Conntents\nError Details: ', detail,'\nFile Name: ',temp);
            except FileNotFoundError as detail:
                print('You have reached the end of your directory\nError Details: ', detail,'\nFile Name: ',temp,'\n\n');
                break;
```
